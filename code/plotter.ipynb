```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
plt.rcParams['figure.figsize'] = [8, 5]
plt.rcParams['figure.dpi'] = 200
colors = list(mcolors.TABLEAU_COLORS.values())
sns.set_theme()
```

```python
def plot_figs(titles, ytitle, prefix="logs/run_2.csv"):
    df = pd.read_csv(prefix, index_col="epoch")
    xs = df.index
    for i, column in enumerate(titles):
        c_train = df[column + " train"].to_numpy()
        c_test = df[column + " test"].to_numpy()
        plt.plot(
            xs[np.isfinite(c_train)], 
            c_train[np.isfinite(c_train)], 
            alpha=0.3,
            color=colors[i],
        )
        plt.plot(
            xs[np.isfinite(c_test)], 
            c_test[np.isfinite(c_test)], 
            label=column,
            color=colors[i],
        )
    plt.legend(prop={'size': 12})
    plt.xlabel("Epoch")
    plt.ylabel(ytitle)
```

```python
titles = [
  "5 neurons accuracy",
  "15 neurons accuracy",
  "30 neurons accuracy",
  "60 neurons accuracy",
  "120 neurons accuracy",
]
plot_figs(titles, "Conditional Entropy")
```

```python
titles = [
  "15 neurons H(Y|X)",
  "15 neurons H(Z|X)",
]
plot_figs(titles, "Conditional Entropy")
```

```python
titles = [
  "15 neurons I(Y,X)",
  "15 neurons I(Z,X)",
]
plot_figs(titles, "Mutual Information", "logs/run_1.csv")
```
